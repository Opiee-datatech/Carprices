```python
from sklearn.metrics import classification_report
y_true = [0, 1, 2, 2, 2]
y_pred = [0, 0, 2, 2, 1]
target_names = ['class 0', 'class 1', 'class 2']
print(classification_report(y_true, y_pred,
target_names=target_names))
```

I detta data settet är där totalt 5 samples så ett väldigt litet dataset. Modellen fick 60% av prediktionerna rätt. 
Klass 0 har 50 % av prediktionerna rätt och recallen visar att den har hittat alla exempel av klass 0. 
men där var bara ett exempel. Modellen mislyckades helt med klass 1. 

Macro avg beytder att alla klasser väger lika mycket. Och visar att modellen är oblanserad. 
Weighted avg innebär att varje klass vägs efter hur många exempel den har i datasetet. Därför påverkar klass 2 resultatet mest, eftersom den har flest samples.

En klassificeringsrapport är en sammanfattning av hur bra en klassificeringsmodell presterar. Och används mycket i maskininlärning. 
precision visar hur många av modellens prediktioner är rätt
recall visar hur många faktiska exempel som hittats.
F1 score är medelvärdet på dessa. 
support är antal samples i varje klass i datasetet.

Den hjälper till och förstå om modellen är obalanserad eller fungerar bra per klass och visar om modellen missar vissa exempel.
